```python
from google.colab import drive
drive.mount('/content/drive')

def string_match_brute(A,P): 
    suma = 0
    for i in range(len(A)-len(P)+1):
        for j in range(len(P)):
            if P[j].capitalize() == A[j+i].capitalize() :
                pass
            else:
                break
        if j+1 == len(P) and A[j+i].capitalize() == P[j].capitalize() :
            suma = suma + 1
            print(f"{P} encontrado en el indice {i}")
    print(f"----> {P} se encontro {suma} veces <----\n")
    return suma 

arch = open("drive/MyDrive/Colab_arch/el_quijote_ii.txt","r", encoding='utf-8')
data = arch.read()
suma = []
suma.append(string_match_brute(data, "molino"))
suma.append(string_match_brute(data, "hidalgo"))
suma.append(string_match_brute(data,"merced"))
arch.close()

print("*** Resumen ***")
print(f"Molino aparece {suma[0]} veces")
print(f"Hidalgo aparece {suma[1]} veces")
```
